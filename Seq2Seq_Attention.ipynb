## Arrange

```python
def seq_data(x, y, xseq_len, yseq_len):  # N=100, in=5, out=3

    xseqlist = []
    yseqlist = []
    it = (len(x) - xseq_len) // yseq_len # Î™´

    for i in range(it):
        xseqlist.append(x[i*yseq_len: i*yseq_len + xseq_len])
        yseqlist.append(y[xseq_len + i*yseq_len: xseq_len + i*yseq_len + yseq_len])
        # e.g.
        # x[0*3: 0*3 + 5]        =x[0:5],     x[1*3: 1*3 + 5]        =x[3:8] 
        # y[5 + 0*3: 5 + 0*3 + 3]=y[5:8],     y[5 + 1*3: 5 + 1*3 + 3]=y[8:11]
        #
        # x[92:97]
        # y[97:100]

    return torch.tensor(xseqlist, dtype=torch.float32).to(device), torch.tensor(yseqlist, dtype=torch.float32).to(device)
```

## __üü®Encoderüü®__
- ÏûÖÎ†• seqÎ•º Í≥†Ï†ïÎêú ÌÅ¨Í∏∞Ïùò Î≤°ÌÑ∞Î°ú Î≥ÄÌôò.
- ÏûÖÎ†• ÏãúÌÄÄÏä§Ïùò Í∞Å ÏãúÍ∞Ñ Ïä§ÌÖùÎßàÎã§ RNNÏùÄ Ìï¥Îãπ ÏãúÍ∞Ñ Ïä§ÌÖùÏùò ÏûÖÎ†•Í≥º Ïù¥Ï†Ñ ÏãúÍ∞Ñ Ïä§ÌÖùÏóêÏÑúÏùò hidden ÏÉÅÌÉúÎ•º ÏÇ¨Ïö©ÌïòÏó¨ hidden ÏÉÅÌÉúÏôÄ cell ÏÉÅÌÉúÎ•º Í∞±Ïã†Ìï©ÎãàÎã§.

- outÏùÄ LSTM Î†àÏù¥Ïñ¥Ïùò Î™®Îì† ÏãúÍ∞Ñ Ïä§ÌÖùÏóêÏÑúÏùò Ï∂úÎ†•ÏùÑ Îã¥Í≥† ÏûàÎäî ÌÖêÏÑúÏûÖÎãàÎã§. LSTMÏùÄ ÏãúÌÄÄÏä§Î•º ÏùΩÏúºÎ©¥ÏÑú Í∞Å ÏãúÍ∞Ñ Ïä§ÌÖùÎßàÎã§ Ìï¥Îãπ Ïä§ÌÖùÏùò Ï∂úÎ†•ÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§. Îî∞ÎùºÏÑú outÏùÄ (batch_size, time_steps, hidden_size) ÌÅ¨Í∏∞Ïùò ÌÖêÏÑúÏù¥Î©∞, time_steps Í∞úÏàòÏùò ÏãúÍ∞Ñ Ïä§ÌÖùÏóê Ìï¥ÎãπÌïòÎäî Ï∂úÎ†•Îì§ÏùÑ Î™®Îëê Ìè¨Ìï®Ìï©ÎãàÎã§. Í∞Å ÏãúÍ∞Ñ Ïä§ÌÖùÏóêÏÑúÏùò hiddenÏùÑ Î™®Îëê Î™®ÏùÄ Í≤É!

```python
# out, (hn, cn) = self.lstm
# out: containing the output features (h_t) from the last layer of the LSTM, for each t.
# hn:  containing the final hidden state for each element in the sequence.
# cn:  containing the final cell state for each element in the sequence.
# Ï∂úÏ≤ò - ÌååÏù¥ÌÜ†Ïπò Í≥µÏãù Î¨∏ÏÑú

class Encoder(nn.Module):
    def __init__(self, input_len, hidden_len, num_layers):
        super(Encoder, self).__init__()
        self.input_len = input_len
        self.hidden_len = hidden_len
        self.num_layers = num_layers
        self.lstm = nn.LSTM(self.input_len, self.hidden_len, self.num_layers, batch_first=True)

    # (Î∞∞ÏπòÏÇ¨Ïù¥Ï¶à, 5Ïä§ÌÖù, 6Í∞ú ÌîºÏ≥ê)
    # x: (32, 5, 6)
    def forward(self, x):
        # h0: (2, 32, 6) 8Í∞úÏßúÎ¶¨ Îã®ÏùºÎ°ú 32Í∞ú, two layer
        h0 = torch.zeros((self.num_layers, x.shape[0], self.hidden_len)).to(device)
        c0 = torch.zeros((self.num_layers, x.shape[0], self.hidden_len)).to(device)

        # self.hcÎäî hiddenÍ≥º cell stateÏùò tupleÏûÑ.
        # outÏùÄ 6Ï¥àÏóê ÎåÄÌïú Í∞í.
        self.lstm.flatten_parameters()
        out, self.hc = self.lstm(x, (h0, c0))        

        # out: (32, 5, 6)       ----> Îß§ timestepÏóê(Ï¥ù 5Í∞ú) 8ÌÅ¨Í∏∞ hiddenÏùÑ ÎΩëÏïÑÎÇ¥Ï§å.
        # self.hc[0]: (2, 32, 6)----> ÎßàÏßÄÎßâ timestepÏùò hidden. Í∞Å 5ÏùºÏπòÏóêÏÑú ÎßàÏßÄÎßâ Í≤ÉÎßå ÎΩëÏïÑÎÇ¥ÎãàÍπå 32Í∞ú.
        #                             two layerÎãàÍπå 2Í∞Ä Îê®.
        return out, self.hc
```

## __üü®Decoderüü®__
- EncoderÎ°úÎ∂ÄÌÑ∞ Î∞õÏùÄ Ïù∏ÏΩîÎî© Î≤°ÌÑ∞(hidden)ÏôÄ Ïù¥Ï†Ñ ÏãúÏ†êÏùò Ï∂úÎ†•(ÏòàÏ∏°)ÏùÑ ÏûÖÎ†•ÏúºÎ°ú Î∞õÏïÑ Ï∂úÎ†• ÏãúÌÄÄÏä§Î•º ÏÉùÏÑ±ÌïòÎäî Ïó≠Ìï†ÏùÑ Ìï©ÎãàÎã§.
- LSTM ÎòêÎäî GRUÏôÄ Í∞ôÏùÄ RNN Í≥ÑÏó¥Ïùò ÎÑ§Ìä∏ÏõåÌÅ¨Î•º ÌÜµÌï¥ Ïù¥ ÏûÖÎ†•Îì§ÏùÑ Ï≤òÎ¶¨ÌïòÏó¨ ÌòÑÏû¨ ÏãúÏ†êÏùò Ï∂úÎ†•ÏùÑ ÏòàÏ∏°Ìï©ÎãàÎã§.
- Ïù¥ ÏòàÏ∏°Îêú Ï∂úÎ†•ÏùÄ Îã§Ïùå ÏãúÏ†êÏùò ÏûÖÎ†•ÏúºÎ°ú ÏÇ¨Ïö©Îê©ÎãàÎã§. Ï¶â, DecoderÎäî Ïù¥Ï†Ñ ÏãúÏ†êÏùò Ï∂úÎ†•ÏùÑ ÌòÑÏû¨ ÏãúÏ†êÏùò ÏûÖÎ†•ÏúºÎ°ú ÏÇ¨Ïö©ÌïòÏó¨ ÏãúÌÄÄÏä§Î•º Ï†êÏßÑÏ†ÅÏúºÎ°ú ÏÉùÏÑ±Ìï©ÎãàÎã§.
- Ïù¥Î†áÍ≤å ÏòàÏ∏°Îêú 6Ï¥àÏùò Ï∂úÎ†•ÏùÑ Îã§Ïùå ÏãúÍ∞Ñ Ïä§ÌÖùÏùò ÏûÖÎ†•ÏúºÎ°ú ÏÇ¨Ïö©ÌïòÏó¨ 7Ï¥àÏóê Ìï¥ÎãπÌïòÎäî Ï∂úÎ†•ÏùÑ ÏòàÏ∏°Ìï©ÎãàÎã§.

```python
class Decoder(nn.Module):
    def __init__(self, input_len, hidden_len, num_layers, output_len):
        super(Decoder, self).__init__()
        self.input_len = input_len
        self.hidden_len = hidden_len
        self.num_layers = num_layers
        self.output_len = output_len
        self.lstm = nn.LSTM(self.input_len, self.hidden_len, self.num_layers, batch_first=True)
        self.linear = nn.Linear(self.hidden_len, self.output_len) # self.input_size ÏÇ¨Ïö©.

    # encoder_hc: (2, 32, 6) ----> (h0, c0)Ïùò Ïó≠Ìï†
    #                              8Í∞úÏßúÎ¶¨ Îã®ÏùºÎ°ú 32Í∞ú, two layer
    def forward(self, x, encoder_hc):
        # x: (32, 1, 6)
        self.lstm.flatten_parameters()
        out, self.hc = self.lstm(x, encoder_hc)
        # out:    (32, 1, 6)
        # self.h: (2, 32, 6)
        # self.c: (2, 32, 6)
        out = self.linear(out)
        # out:    (32, 1, 6) ----> DecoderÏóê Îã§Ïãú ÎÑ£Í∏∞ ÏúÑÌï¥ inputÍ≥º Í∞ôÏùÄ shapeÏùÑ Í∞ÄÏ†∏Ïïº Ìï®.
        return out
```

## __üü®Karinaüü®__
- 1. ÏµúÏ¢Ö ÏòàÏ∏° Ï∂úÎ†•Îã®ÏóêÏÑú output_seq Í∏∏Ïù¥Ïóê ÎßûÍ≤å reduction Ìï¥Ï§å.
- 2. Shape ÎßûÏ∂∞Ï£ºÎäî Ïö©ÎèÑÎ°ú ÏÇ¨Ïö©.

```python
class Karina(nn.Module):
    def __init__(self, input_len, hidden_len, output_len):
        super(Karina, self).__init__()
        self.input_len = input_len
        self.hidden_len = hidden_len
        self.output_len = output_len
        self.linear1 = nn.Linear(self.input_len, self.hidden_len)
        self.linear2 = nn.Linear(self.hidden_len, self.output_len)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        x = self.relu(x)
        return x
```

## __üü®Attentionüü®__
- EncoderÏùò Ï†ÑÏ≤¥ hiddenÍ≥º DecoderÏùò tÏãúÍ∞Ñ hiddenÏùÑ ÎÇ¥Ï†ÅÌï®.(Attention Score)
- ÎÇ¥Ï†ÅÌïú Í≤∞Í≥ºÎ•º nn.Softmax() ÌÜµÍ≥º.(Attention Distribution; Weight)
- nn.Softmax() ÌÜµÍ≥ºÌïú Í≤∞Í≥ºÎ•º Encoder Ï†ÑÏ≤¥ hiddenÍ≥º Element-wise multiplication.
- ÏúÑ Í≤∞Í≥ºÎ•º FCNÏóê ÎÑ£Ïùå.
- ÏúÑ Í≤∞Í≥ºÎ•º nn.Tanh()Ïóê ÎÑ£Ïùå.

```python
# ÏïÑÎûò ÌÖêÏÑú Ïç®ÏÑú ÌÖåÏä§Ìä∏.
#enc = torch.tensor(np.random.randint(1,8, 32*5*6), dtype=torch.float32).view((32, 5, 6))
#dec = torch.tensor(np.random.randint(1,4, 32*1*6), dtype=torch.float32).view((32, 1, 6))
#print('enc', enc.shape)
#print('dec', dec.shape)


class Attention(nn.Module):
    def __init__(self, feature_len):
        super(Attention, self).__init__()
        self.linear = nn.Linear(feature_len, feature_len)
        self.tanh = nn.Tanh()
        self.softmax = nn.Softmax(dim=1)#----> dim=1Ïù¥ 5ÏùºÏπòÏù¥ÎØÄÎ°ú, 5ÏùºÏπò Îã§ ÎçîÌï¥ÏÑú 1.0 ÎÇòÏò§Í≤å ÎßåÎìúÎäî Í±∞ÏûÑ.


    def forward(self, encout, decout):
        self.encout = encout
        self.decout = decout
        # self.encout: (32, 5, 6)
        # self.decout: (32, 1, 6)

        att_score = torch.matmul(encout, decout.transpose(1,2))# ----> (matrix multiplication)
        # self.decout.transpose(1, 2): (32, 6, 1)
        # score: (32, 5, 1)
        # ÌïòÎ£®Ïóê Ìïú Í∞úÏî©(1) 5Ïùº Í∞Ñ(5)Ïùò attention scoreÍ∞Ä Ï∂úÎ†•Îê®. Í∑∏Í≤å 32Í∞ú Î∞∞ÏπòÎ°ú.

        att_dist = self.softmax(att_score)
        # att_dist: (32, 5, 1)

        att_dist = att_dist.expand(-1, -1, 6)
        # att_dist: (32, 5, 6)
        # Ïó∞ÏÇ∞ ÏúÑÌï¥ Î≥µÏÇ¨Ìï¥ÏÑú ÌôïÏû•.

        att_value = torch.mul(encout, att_dist)# ----> (element-wise multiplication)
        # att_value: (32, 5, 6)

        v = torch.cat((att_value, decout), dim=1)
        # v: (32, 6, 6)

        v = self.linear(v) # ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ïù¥Î†áÍ≤å Ï†ÅÏö©. Ïù¥Ïú†Îäî Î™®Î¶Ñ.
        # v: (32, 6, 6)
        v = self.tanh(v)   # ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ïù¥Î†áÍ≤å Ï†ÅÏö©. Ïù¥Ïú†Îäî Î™®Î¶Ñ.
        # v: (32, 6, 6)

        return v
        # v: (32, 6, 6)
```

## __üü®Seq2Seqüü®__
- encoderÏóêÏÑúÎäî ÏûÖÎ†• ÏãúÌÄÄÏä§Î•º Î∞õÏïÑ context vector(hidden, cell)Î•º decoderÏóê Ï†ÑÎã¨.
- decoderÎäî context vectorÎ•º ÏûÖÎ†• Î∞õÏïÑ Îã§Ïùå ÏãúÏ†ê Í∞íÏùÑ ÏòàÏ∏°Ìï®. Í∑∏ ÏòàÏ∏°Ïù¥ Îã§Ïãú Îã§Ïùå timestepÏóê ÏûÖÎ†•ÏúºÎ°ú Îì§Ïñ¥Í∞Ä Îòê ÏòàÏ∏°. Ïù¥Í±∏ Î∞òÎ≥µ.

#### Teacher Forcing
- decoderÏùò (t-1)ÏòàÏ∏°Ïù¥ tÏóê inputÏúºÎ°ú Îì§Ïñ¥Í∞ÄÎäî Í≤ÉÏùÄ ÌÖåÏä§Ìä∏Ìï† ÎïåÏùò ÏÇ¨Ïö©Î≤ïÏûÑ!
- teacher forcingÏùÄ (t-1)ÏùÑ tÏóê inputÏúºÎ°ú ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÍ≥† t ÏãúÏ†ê Ïã§Ï†ú Í∞íÏùÑ ÏûÖÎ†•ÏúºÎ°ú ÎÑ£Ïñ¥ ÌïôÏäµÏãúÌÇ¥!

#### cons
- context vector ÎßåÎì§Î©¥ÏÑú Ï†ïÎ≥¥ ÏÜêÏã§.
- ÏûÖÎ†• Í∏∏Ïñ¥ÏßÄÎ©¥ ÏÑ±Îä• Ï†ÄÌïò.

#### ÏãúÌñâ Ï∞©Ïò§
- shapeÏù¥ Ïïà ÎßûÎäî Í≤ΩÏö∞, 2dimÏúºÎ°ú Î∞îÍæº Îí§, nn.LinearÏóê ÎÑ£Ïñ¥ÏÑú Í∞ïÏ†úÎ°ú ÎßûÏ∂∞ Ï£ºÏóàÏùå. ---> ÏïåÍ≥†Î¶¨Ï¶òÏù¥ Î¨¥Í±∞ÏõåÏßÄÍ≤†ÏßÄÎßå ÏûëÎèôÏùÄ Ïûò Ìï®.
- shapeÏù¥ Ïïà ÎßûÎäî Í≤ΩÏö∞, ÏùºÎ∂ÄÎßå ÎñºÏñ¥ÎÇ¥ÏÑú Í∞ïÏ†úÎ°ú shapeÏùÑ ÎßûÏ∂∞Î¥§Ïùå. ----> ÌçºÌè¨Î®ºÏä§ Ïïà Ï¢ãÏùå. Ï†ïÎ≥¥ ÏÜêÏã§Ïù∏ ÎìØ. (Ïïà ÏîÄ.)

```python
class Seq2Seq(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, xseq_len, yseq_len, batch_len):
        super(Seq2Seq, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.output_size = output_size
        self.xseq_len = xseq_len
        self.yseq_len = yseq_len
        self.batch_len = batch_len

        self.attention = Attention(feature_len=self.input_size)
        self.forpred = Karina(input_len=self.input_size*self.input_size*self.yseq_len, hidden_len=64, output_len=self.yseq_len)
        self.forreduction = Karina(input_len=self.input_size*self.input_size, hidden_len=64, output_len=self.input_size)
        self.encoder = Encoder(self.input_size, self.hidden_size, self.num_layers)
        self.decoder = Decoder(self.input_size, self.hidden_size, self.num_layers, self.input_size)

    # y_dataÏôÄ teacher_probÎäî teacher forcingÌï† Îïå ÎÑ£Îäî Í±∞ÏûÑ.
    def forward(self, x_data): 
        # x_data: (32, 5, 6)
        # y_data: (32, 4, 1)
        encoder_out, encoder_hc = self.encoder(x_data)
        # encoder_out:  (32, 5, 6)
        # encoder_hc[0]:(2, 32, 6)  
        # encoder_hc[1]:(2, 32, 6)  

        decoder_in = x_data[:, -1, :].unsqueeze(1)
        # decoder_init_in: (32, 6).unsqueeze(1) = (32, 1, 6)
        # feature 6Í∞úÏßúÎ¶¨Î°ú 5Ïùº Ï§ë ÎßàÏßÄÎßâ ÎÇ†Îßå ÎΩëÏùå. Í∑∏Í≤å 32Í∞ú Batch.
        # Decoder ÌÅ¥ÎûòÏä§ÏóêÏÑú hidden+cell Ï†ÑÎã¨ÌïòÎ©¥ÏÑú Í∞ôÏù¥ Ï†ÑÎã¨Ìï®.

        decoder_in = self.attention(encoder_out, decoder_in)
        # Attention Ï†ÅÏö©.
        # decoder_in: (32, 6, 6)

        box = torch.zeros((self.batch_len, 1), dtype=torch.float32).to(device)
        # box: (32, 1) ----> Ï¥àÍ∏∞ Í∞í.
        # DecoderÏùò outputÎì§ÏùÑ ÏåìÏïÑÏÑú Ï†ÄÏû•Ìï¥ ÎÜìÏùå.

        for t in range(self.yseq_len):
            decoder_out = self.decoder(decoder_in, encoder_hc)
            # decoder_out: (32, 6, 6)

            decoder_out_att = decoder_out.view(self.batch_len, -1).unsqueeze(1)
            # decoder_out_att: (32, 1, 6)
            decoder_out_att = self.forreduction(decoder_out_att)
            # decoder_out_att: (32, 1, 6)

            decoder_out_nextin = self.attention(encoder_out, decoder_out_att)
            # decoder_out_nextin: (32, 6, 6)

            decoder_out_store = decoder_out.view(self.batch_len, -1)
            # decoder_out_forstore: (32, 6*6)
            box = torch.cat([box, decoder_out_store], axis=1)

            decoder_in = decoder_out_nextin
            # decoder_in: (32, 6, 6)


        box = box[:, 1:]
        # box: (32, 6*6*4)

        y_pred = self.forpred(box)
        # y_pred: (32, 4)
        y_pred = y_pred.unsqueeze(2)
        # y_pred: (32, 4, 1)
        
        return y_pred
```

## __üü®Trainingüü®__

```python
model = Seq2Seq(input_size=infeature_len, hidden_size=6, num_layers=2, output_size=infeature_len,
                xseq_len=xseq_len, yseq_len=yseq_len, batch_len=BATSIZE).to(device)
criterion = nn.MSELoss().to(device)
optimizer = optim.RAdam(model.parameters(), lr=1e-3)
```

```python
loss_graph = []
n = len(trainloader)

for epoch in tqdm(range(1, 800+1)):
    running_loss = 0.0
    
    for x_true, y_true in trainloader:
        
        y_pred = model(x_true)
        loss = criterion(y_pred, y_true)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    
    loss_graph.append(running_loss/n)
    if epoch % 100 == 0:
        print('[epoch: %d] loss: %.4f' %(epoch, running_loss/n))
```
